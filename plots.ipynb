```python
import os
import joblib
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_dir = r'C:\Users\jnicolow\Documents\courses\fall2024\ICS-661\final project\sampleDataset'

original_raw_list = joblib.load(os.path.join(data_dir, 'ocr_cosine_sim_list_ocr_output.pkl'))
original_gpt_list = joblib.load(os.path.join(data_dir, 'gpt_cosine_sim_list_ocr_output.pkl'))

scale4_raw_list = joblib.load(os.path.join(data_dir, 'ocr_cosine_sim_list_ocr_output_scalefactor4.pkl'))
scale4_gpt_list = joblib.load(os.path.join(data_dir, 'gpt_cosine_sim_list_ocr_output_scalefactor4.pkl'))

scale6_raw_list = joblib.load(os.path.join(data_dir, 'ocr_cosine_sim_list_ocr_output_scalefactor6.pkl'))
scale6_gpt_list = joblib.load(os.path.join(data_dir, 'gpt_cosine_sim_list_ocr_output_scalefactor6.pkl'))

```

```python
raw_means = [
    np.mean(original_raw_list),  # Original (0 scale)
    np.mean(scale4_raw_list),    # Scale factor 4
    np.mean(scale6_raw_list)     # Scale factor 6
]

gpt_means = [
    np.mean(original_gpt_list),  # Original (0 scale)
    np.mean(scale4_gpt_list),    # Scale factor 4
    np.mean(scale6_gpt_list)     # Scale factor 6
]

# Step 2: Prepare x-axis values (scale factors)
x_values = [0, 4, 6]

# Step 3: Plot the data
plt.figure(figsize=(8, 6))

# Plot raw vs GPT means
plt.plot(x_values, raw_means, label="Raw", marker='o', color='blue')
# plt.plot(x_values, gpt_means, label="GPT 3.5 clean", marker='o', color='red')

# Step 4: Add labels and title
plt.xlabel("Scale Factor")
plt.ylabel("Mean Cosine Similarity")
plt.title("Cosine Similarity Means vs Scale Factor")
plt.xticks(x_values)  # Ensure x-ticks match the scale factor values

# Step 5: Add a legend
# plt.legend()

# Step 6: Show the plot
plt.grid(False)
plt.show()
```
