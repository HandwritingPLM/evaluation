```python
import os
import joblib
import matplotlib.pyplot as plt
from matplotlib import cm  # For color maps

import numpy as np
```

```python
data_dir = r'output embedding sims'

# This is the same for everything so can just choose arbitrarily
prompt = 'prompt1'
original_raw_list = joblib.load(os.path.join(data_dir, prompt, 'ocr_cosine_sim_list_ocr_output.pkl'))
scale4_raw_list = joblib.load(os.path.join(data_dir, prompt, 'ocr_cosine_sim_list_ocr_output_scalefactor4.pkl'))
scale6_raw_list = joblib.load(os.path.join(data_dir, prompt, 'ocr_cosine_sim_list_ocr_output_scalefactor6.pkl'))

raw_means = [
    np.mean(original_raw_list),  # Original (0 scale)
    np.mean(scale4_raw_list),    # Scale factor 4
    np.mean(scale6_raw_list)     # Scale factor 6
]



# Step 2: Prepare x-axis values (scale factors)
x_values = [0, 4, 6]

# Step 3: Plot the data
plt.figure(figsize=(8, 6))
color_map = cm.get_cmap('tab10', 4)  # 'tab10' is a colormap with 10 distinct colors


# Plot raw vs GPT means
plt.plot(x_values, raw_means, label="Raw", marker='o', color='blue')



prompt_dict = {'prompt1':'Zero Shot 1', 'prompt2': 'Zero Shot 2', 'prompt3':'One Shot', 'prompt4':'Few Shot'}


# for the raw just pick one
for i, prompt in enumerate(os.listdir(data_dir)):
    print(prompt)
    original_gpt_list = joblib.load(os.path.join(data_dir, prompt, 'gpt_cosine_sim_list_ocr_output.pkl'))
    scale4_gpt_list = joblib.load(os.path.join(data_dir, prompt, 'gpt_cosine_sim_list_ocr_output_scalefactor4.pkl'))
    scale6_gpt_list = joblib.load(os.path.join(data_dir, prompt, 'gpt_cosine_sim_list_ocr_output_scalefactor6.pkl'))
    gpt_means = [
        np.mean(original_gpt_list),  # Original (0 scale)
        np.mean(scale4_gpt_list),    # Scale factor 4
        np.mean(scale6_gpt_list)     # Scale factor 6
    ]
    gpt_stds = [
        np.std(original_gpt_list),  
        np.std(scale4_gpt_list),    
        np.std(scale6_gpt_list)     
    ]

    plt.plot(x_values, gpt_means, label=f"{prompt_dict[prompt]} GPT 3.5 clean", marker='o', color=color_map(i))

    # plt.errorbar(
    #     x_values, 
    #     gpt_means, 
    #     yerr=gpt_stds, 
    #     label=f"{prompt_dict[prompt]} GPT 3.5 clean", 
    #     marker='o', 
    #     color=color_map(i), 
    #     capsize=5  # Adds caps to the error bars
    # )

    
    
# Step 4: Add labels and title
plt.xlabel("Scale Factor")
plt.ylabel("Mean Cosine Similarity")
plt.title("Cosine Similarity Means vs Scale Factor")
plt.xticks(x_values)  # Ensure x-ticks match the scale factor values

# Step 5: Add a legend
plt.legend()

# Step 6: Show the plot
plt.grid(False)
plt.show()



```

```python
from glob import glob

glob(os.path.join('output embedding sims', '*', 'gptcleaned-ocr_output', '*.txt'))

file_pattern = os.path.join('output embedding sims', 'prompt4', 'gptcleaned_ocr_output', '*.txt')

# Get all matching text file paths
file_paths = glob(file_pattern)

# Initialize a list to store the lengths of the files
file_lengths = []

# Iterate through each file, read its content, and calculate its length
for file_path in file_paths:
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.read()
        file_lengths.append(len(content))

# Calculate the min, median, and max lengths
min_length = np.min(file_lengths)
median_length = np.median(file_lengths)
max_length = np.max(file_lengths)

# Output the results
print(f"Minimum length: {min_length}")
print(f"Median length: {median_length}")
print(f"Maximum length: {max_length}")
```
